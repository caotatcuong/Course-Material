```python
import pandas as pd
import glob
import re
```

```python
list_pd = []
for df_path in glob.glob("/content/Data agri" + "/*.csv"):
  df= pd.read_csv(df_path,header = 1).set_index('Tỉnh, thành phố').T.reset_index().rename(columns = {'index': 'Năm'}).rename_axis(None, axis=1)
  with open(df_path) as f:
    df.name = re.findall(r'"([^"]*)"',f.readline())[0]
  list_pd.append(df)
```
